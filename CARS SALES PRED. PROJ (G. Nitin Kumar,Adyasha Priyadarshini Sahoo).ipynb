```python
#ip/op creation
ip=cars.drop('Sales',axis=1)
op=cars.Sales
```

```python
ip.head()
```

```python
op.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
#ML model
#linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
y_test
```

```python
#Accuracy
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(pred,y_test)
r2=r2_score(pred,y_test)

print('Mean Square Error: ',mse)
print('r2 Score: ',r2)
```
